```python
import pandas as pd

df = pd.read_csv("Resources/vine_table.csv")
df 
```

```python
# Filter and retrieve all the rows where the total_votes count is equal to or greater than 20
high_vote_df = df[df["total_votes"] >= 20]
high_vote_df
```

```python
# Filter the new DataFrame and create a new DataFrame to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%
helpful_votes_df = high_vote_df[(high_vote_df["helpful_votes"] / high_vote_df["total_votes"]) >= .50]
helpful_votes_df
```

```python
# Filter the DataFrame from the previous step and create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'
helpful_vines_df = helpful_votes_df[helpful_votes_df["vine"] == "Y"]
helpful_vines_df
```

```python
# Repeat the previous step, but this time retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'
helpful_non_df = helpful_votes_df[helpful_votes_df["vine"] == "N"]
helpful_non_df
```

```python
# Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
total_reviews = len(helpful_non_df)
five_stars = len(helpful_non_df[helpful_non_df["star_rating"] == 5])
percentage = (five_stars / total_reviews * 100)
```

```python
print(f"Total # of non-Vine reviews:{total_reviews}")
print(f"# of 5 star reviews:{five_stars}")
print(f"% of 5 star reviews:{percentage}%")
```
